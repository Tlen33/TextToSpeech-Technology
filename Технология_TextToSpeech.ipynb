Установим библиотеку gTTS и с помощью неё синтезируем аудио следующего текста:

    Многие знают много, но никто не знает всё

Сделаем синтез на трёх языках: 
*   русском (ru)
*   английском (en)
*   немецком (de)


```python
%%capture

# Бесплатная библиотека Google для задачи TTS
!pip install gTTS
```

```python
# Импорт библиотек

import gtts 
from gtts import gTTS 
from IPython.display import Audio 
```

**Синтез русского текста на русском языке**

```python
# Зададим текст
text = """Многие знают много, но никто не знает всё"""

# Преобразуем текст в аудио. Параметр lang='ru' - русский язык
tts = gTTS(text, lang='ru') 

# Сохраняем в формате мр3
tts.save('russ.mp3') 
```

```python
# Прослушаем полученный результат
Audio('russ.mp3')
```

**Синтез русского текста с параметром сети английского языка lang='en'**

```python
# Зададим текст
text = """Многие знают много, но никто не знает всё"""

# Преобразуем текст в аудио. Параметр lang='en' - английский язык
tts = gTTS(text, lang='en') 

# Сохраняем в формате мр3
tts.save('eng.mp3') 
```

```python
# Прослушаем полученный результат
Audio('eng.mp3')
```

**Синтез русского текста с параметром сети немецкого языка lang='de'**

```python
# Зададим текст
text = """Многие знают много, но никто не знает всё"""

# Преобразуем текст в аудио. Параметр lang='de' - украинский язык
tts = gTTS(text, lang='de') 

# Сохраняем в формате мр3
tts.save('nem.mp3') 
```

```python
# Прослушаем полученный результат
Audio('nem.mp3')
```
